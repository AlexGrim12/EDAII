```python
# Gutierrez Grimaldo Alejandro
# Definición de la función para insertar un par llave-valor en la tabla hash.
def insertar(T, m, x, valor):
    j = 0
    # Calcular el hash de la llave x
    h1 = h(convertirLlave(x), m)
    while j < m:
        # Calcular el índice de inserción en la tabla hash
        indice = (h1 + j) % m
        par = [x, valor]
        if T[indice] == None:
            # Si el índice está vacío, insertar el par llave-valor y devolver el índice.
            T[indice] = par
            return indice
        else:
            j += 1
    # Si no se encuentra un lugar vacío después de recorrer la tabla, imprimir un mensaje de error.
    print("No hay lugar o la llave está repetida")
    return -1

# Función para crear una tabla hash vacía de tamaño m.
def formarTabla(m):
    T = [None] * m
    return T

# Función para convertir una llave RFC en un número.
def convertirLlave(x):
    keyNum = 0
    i = 0
    for char in x:
        # Sumar los valores ponderados de los caracteres de la llave.
        keyNum += ord(char) * i
        i += 1
    return keyNum

# Función de hash simple que toma el residuo del número convertido entre m.
def h(x, m):
    return x % m

# Función principal del programa.
def main():
    m = 11
    # Crear una tabla hash vacía.
    T = formarTabla(m)
    for i in range(2):
        x = input("Ingrese el RFC: ")
        nombre = input("Ingrese el nombre: ")
        edad = input("Ingrese la edad: ")
        salario = input("Ingrese el salario: ")
        valor = nombre + " " + edad + " " + salario
        # Insertar el par llave-valor en la tabla hash.
        insertar(T, m, x, valor)
    # Mostrar la tabla hash resultante.
    print(T)

# Llamar a la función principal para ejecutar el programa.
main()
```
