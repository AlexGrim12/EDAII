```python
class Nodo:
    def __init__(self, llave):
        self.hijoIzq = None
        self.hijoDer = None
        self.padre = None
        self.llave = llave
        
class Arbol:
    def __init__(self):
        self.raiz = None
    def obtenerRaiz(self):
        return self.raiz

    def insertar(self, llave):
        if self.raiz == None:
            self.raiz = Nodo(llave)
        else:
            self._insertar(llave, self.raiz)
            
    def _insertar(self, llave, nodoActual):
        if llave < nodoActual.llave:
            if nodoActual.hijoIzq == None:
                nodoActual.hijoIzq = Nodo(llave)
                nodoActual.hijoIzq.padre = nodoActual
            else:
                self._insertar(llave, nodoActual.hijoIzq)
        elif llave > nodoActual.llave:
            if nodoActual.hijoDer == None:
                nodoActual.hijoDer = Nodo(llave)
                nodoActual.hijoDer.padre = nodoActual
            else:
                self._insertar(llave, nodoActual.hijoDer)
        else:
            print("El nodo ya existe en el arbol")      

    def inorden(self):
        if self.raiz != None:
            self._inorden(self.raiz)
            
    def _inorden(self, nodoActual):
        if nodoActual != None:
            self._inorden(nodoActual.hijoIzq)
            print(str(nodoActual.llave) + ' ')
            self._inorden(nodoActual.hijoDer)
            
    def preorden(self):
        if self.raiz != None:
            self._preorden(self.raiz)
            
    def _preorden(self, nodoActual):
        if nodoActual != None:
            print(str(nodoActual.llave) + ' ')
            self._preorden(nodoActual.hijoIzq)
            self._preorden(nodoActual.hijoDer)
            
    def postorden(self):
        if self.raiz != None:
            self._postorden(self.raiz)
            
    def _postorden(self, nodoActual):
        if nodoActual != None:
            self._postorden(nodoActual.hijoIzq)
            self._postorden(nodoActual.hijoDer)
            print(str(nodoActual.llave) + ' ')      
                
    def buscar(self, llave):
        if self.raiz != None:
            return self._buscar(llave, self.raiz)
        else:
            return None
        
    def _buscar(self, llave, nodoActual):
        if llave == nodoActual.llave:
            return nodoActual
        elif llave < nodoActual.llave and nodoActual.hijoIzq != None:
            return self._buscar(llave, nodoActual.hijoIzq)
        elif llave > nodoActual.llave and nodoActual.hijoDer != None:
            return self._buscar(llave, nodoActual.hijoDer)
        elif llave != nodoActual.llave:
            return None
        
# imprimir el camino desde un nodo hasta la raiz
    def caminoHastaRaiz(self, nodo):
        if nodo != None:
            print(str(nodo.llave) + ' ')
            self.caminoHastaRaiz(nodo.padre)
            
    def altura(self):
        if self.raiz != None:
            return self._altura(self.raiz, 0)
        else:
            return 0
        
    def _altura(self, nodoActual, alturaActual):
        if nodoActual == None: return alturaActual
        alturaIzq = self._altura(nodoActual.hijoIzq, alturaActual + 1)
        alturaDer = self._altura(nodoActual.hijoDer, alturaActual + 1)
        return max(alturaIzq, alturaDer)
        
    def minimoArbol(self):
        nodoActual = self.raiz
        while nodoActual.hijoIzq != None:
            nodoActual = nodoActual.hijoIzq
        return nodoActual.llave
        
    def maximoArbol(self):
        nodoActual = self.raiz
        while nodoActual.hijoDer != None:
            nodoActual = nodoActual.hijoDer
        return nodoActual.llave
    
    # Implementar el borrado de un ABB. 
    def borrar(self, llave):
        if self.raiz == None:
            return False
        elif self.raiz.llave == llave:
            if self.raiz.hijoIzq == None and self.raiz.hijoDer == None:
                self.raiz = None
            elif self.raiz.hijoIzq == None:
                self.raiz = self.raiz.hijoDer
            elif self.raiz.hijoDer == None:
                self.raiz = self.raiz.hijoIzq
            else:
                self._borrar(llave, self.raiz)
        else:
            self._borrar(llave, self.raiz)
        
    def _borrar(self, llave, nodoActual):
        if llave < nodoActual.llave:
            if nodoActual.hijoIzq != None:
                self._borrar(llave, nodoActual.hijoIzq)
        elif llave > nodoActual.llave:
            if nodoActual.hijoDer != None:
                self._borrar(llave, nodoActual.hijoDer)
        else:
            if nodoActual.hijoIzq == None and nodoActual.hijoDer == None:
                self._reemplazarNodo(nodoActual, None)
            elif nodoActual.hijoIzq == None:
                self._reemplazarNodo(nodoActual, nodoActual.hijoDer)
            elif nodoActual.hijoDer == None:
                self._reemplazarNodo(nodoActual, nodoActual.hijoIzq)
            else:
                nodoReemplazo = self._obtenerNodoReemplazo(nodoActual)
                if nodoReemplazo != None:
                    self._borrar(nodoReemplazo.llave, nodoReemplazo)
                    nodoActual.llave = nodoReemplazo.llave
                    
    def _reemplazarNodo(self, nodoActual, nodoNuevo):
        if nodoActual.padre != None:
            if nodoActual == nodoActual.padre.hijoIzq:
                nodoActual.padre.hijoIzq = nodoNuevo
            else:
                nodoActual.padre.hijoDer = nodoNuevo
        if nodoNuevo != None:
            nodoNuevo.padre = nodoActual.padre
            
    def _obtenerNodoReemplazo(self, nodoActual):
        nodoReemplazo = None
        if nodoActual != None and nodoActual.hijoDer != None:
            nodoReemplazo = self._obtenerNodoMasIzq(nodoActual.hijoDer)
        return nodoReemplazo
    
    def _obtenerNodoMasIzq(self, nodoActual):
        if nodoActual.hijoIzq == None:
            return nodoActual
        else:
            return self._obtenerNodoMasIzq(nodoActual.hijoIzq)   
                 
def printBuscar(nodo):
    if nodo != None:
        print("Se encontro " + str(nodo.llave))
    else:
        print("No se encontro el nodo")
           
def printCamino(nodo):
    if nodo != None:
        print("El camino desde el nodo hasta la raiz es: ")
        arbol.caminoHastaRaiz(nodo)
    else:
        print("No se encontro el nodo")
        
def printAltura(nodo):
    if nodo != None:
        print("La altura del nodo es: " + str(arbol.altura()))
    else:
        print("No se encontro el nodo")
        
def printMinimo(nodo):
    if nodo != None:
        print("El minimo del arbol es: " + str(arbol.minimoArbol()))
    else:
        print("No se encontro el nodo")
        
def printMaximo(nodo):
    if nodo != None:
        print("El maximo del arbol es: " + str(arbol.maximoArbol()))
    else:
        print("No se encontro el nodo")

arbol = Arbol()
arbol.insertar(10)
arbol.insertar(3)
arbol.insertar(2)
arbol.insertar(18)
arbol.insertar(7)
arbol.insertar(12)
arbol.insertar(7)
arbol.insertar(4)
arbol.insertar(20)
arbol.insertar(0)

print("Se buscaran las llaves 3, 100, 20")
print("Se busca 3")
printBuscar(arbol.buscar(3))
print("Se busca 100")
printBuscar(arbol.buscar(100))
print("Se busca 20")
printBuscar(arbol.buscar(20))

print("Se buscara el camino desde el nodo 3 hasta la raiz")
printCamino(arbol.buscar(3))
print("Se buscara la altura del nodo 3")
printAltura(arbol.buscar(3))
print("Se buscara el minimo del arbol")
printMinimo(arbol.buscar(3))
print("Se buscara el maximo del arbol")
printMaximo(arbol.buscar(3))

print("Se borrara el nodo 3")
arbol.borrar(3)
print("Se buscara el nodo 3")
printBuscar(arbol.buscar(3))
print("Se buscara el camino desde el nodo 3 hasta la raiz")
printCamino(arbol.buscar(3))
print("Se buscara la altura del nodo 3")
printAltura(arbol.buscar(3))
```
