```python
#Gutiérrez Grimaldo Alejandro

def bubbleSort(lista):
    print("Lista original")
    print([lista], end="\n\n")
    for i in range(1, len(lista)):
        print(f"Pasada {i} \n")
        for j in range(len(lista)-1):
            if lista[j]>lista[j+1]:
                temp = lista[j]
                lista[j] = lista[j+1]
                lista[j+1] = temp
            print(lista)
        print()
lista=[8,20,2,39,11,34,22,5,2,0]
bubbleSort(lista)
```

```python
#Gutiérrez Grimaldo Alejandro

def bubbleSortMejorado(lista):
    flag = True
    i = 1
    print("Lista original")
    print([lista], end="\n\n")
    while i < len(lista)-1 and flag:
        print(f"Pasada {i} \n")
        flag = False
        for j in range(len(lista)-1-i):
            if(lista[j]>lista[j+1]):
                flag = True
                temp = lista[j]
                lista[j] = lista[j+1]
                lista[j+1] = temp
            print(lista)
        print()
        i = i + 1
lista=[8,20,2,39,11,34,22,5,2,0]
bubbleSortMejorado(lista)
```

```python
#Gutiérrez Grimaldo Alejandro

def Merge(lista, p, q, r):
    izq = lista[p:(q + 1)]
    der = lista[q+1:(r + 1)]
    print("Izquierda:", izq)
    print("Derecha:", der)
    i = 0
    j = 0
    for k in range(p, r + 1):
        if(j >= len(der) or (i < len(izq) and izq[i] < der[j])):
            lista[k] = izq[i]
            i = i + 1
        else:
            lista[k] = der[j]
            j = j + 1
    print("Combinaciòn:", lista, end="\n\n")
            

def MergeS(lista, p, r):
    if(p < r):
        q = int((p + r)/2)
        MergeS(lista, p, q)
        MergeS(lista, q + 1, r)
        Merge(lista, p, q, r)


        
lista = [8,20,2,39,11,34,22,5,2,0]
print("Lista original")
print([lista], end="\n\n")
MergeS(lista, 0, 9)
```
