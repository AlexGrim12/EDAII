```python
#Gutiérrez Grimaldo Alejandro

import math

def hIzq(i):
    return 2*i + 1

def hDer(i):
    return 2*i + 2

def change(lista, x, y):
    lista[x], lista[y] = lista[y], lista[x]
    
def maxHeapify(lista, i, heapSize):
    l = hIzq(i)
    r = hDer(i)
    if l < heapSize and lista[l] > lista[i]:
        largest = l
    else:
        largest = i
    if r < heapSize and lista[r] > lista[largest]:
        largest = r
    if largest != i:
        change(lista, i, largest)
        maxHeapify(lista, largest, heapSize)
        
def buildMaxHeap(lista):
    heapSize = len(lista)
    for i in range(math.floor(len(lista)/2), -1, -1):
        maxHeapify(lista, i, heapSize)
        
def heapSort(lista):
    buildMaxHeap(lista)
    print("\nHeap Max Inicial: ", lista)
    heapSize = len(lista)
    print("\nEmpieza el ordenamiento: \n")
    for i in range(len(lista)-1, 0, -1):
        print("Intercambio: ", lista[0], "con", lista[i])
        change(lista, 0, i)
        print("\nElemento ", lista[i], "ordenado")
        heapSize = heapSize - 1
        maxHeapify(lista, 0, heapSize)
        print("\nReconstruccion del Heap Max: ", lista[0:heapSize])

lista = [16, 5, 2, 1, 0]
print("*****Heap Sort*****")
print("Lista original: ", lista)
heapSort(lista)
print("\nLista ordenada: ", lista)
```

```python
def hIzq(i):
    return 2 * i + 1

def hDer(i):
    return 2 * i + 2

def change(lista, x, y):
    lista[x], lista[y] = lista[y], lista[x]
    
def minHeapify(lista, i, heapSize):
    l = hIzq(i)
    r = hDer(i)
    if l < heapSize and lista[l] < lista[i]:
        smallest = l
    else:
        smallest = i
    if r < heapSize and lista[r] < lista[smallest]:
        smallest = r
    if smallest != i:
        change(lista, i, smallest)
        minHeapify(lista, smallest, heapSize)
        
def buildMinHeap(lista):
    heapSize = len(lista)
    for i in range(math.floor(len(lista)/2), -1, -1):
        minHeapify(lista, i, heapSize)

def heapSort(lista):
    buildMinHeap(lista)
    print("\nHeap Min Inicial: ", lista)
    heapSize = len(lista)
    print("\nEmpieza el ordenamiento: \n")
    for i in range(len(lista)-1, 0, -1):
        print("Intercambio: ", lista[0], "con", lista[i])
        change(lista, 0, i)
        print("\nElemento ", lista[i], "ordenado")
        heapSize = heapSize - 1
        minHeapify(lista, 0, heapSize)
        print("\nReconstruccion del Heap Min: ", lista[0:heapSize])

def invertir(lista):
    if len(lista) == 0:
        return []
    else:
        return [lista[-1]] + invertir(lista[:-1])
    
lista = [16, 5, 2, 1, 0]
print("*****Heap Sort*****")
print("Lista original: ", lista)
heapSort(lista)
print("\nLista en orden descendente: ", lista)
print("\nLista en orden: ", invertir(lista))


```
