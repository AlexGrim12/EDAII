```python
class Vertice:
    def __init__(self, n):
        self.nombre = n
        self.vecinos = []
    def agregarVecino(self, v):
        if v not in self.vecinos:
            self.vecinos.append(v)
            self.vecinos.sort()
    
class Grafo:
    def __init__(self, tipo):
        self.vertices = {}
        self.tipo = tipo
    def agregarVertice(self, v):
        if v not in self.vertices:
            self.vertices[v] = Vertice(v)
            
    def agregarArista(self, u, v):
        if u in self.vertices and v in self.vertices:
            if self.tipo == "1":
                self.vertices[u].agregarVecino(v)
            else:
                self.vertices[u].agregarVecino(v)
                self.vertices[v].agregarVecino(u)
    def imprimirGrafo(self):
        for v in self.vertices:
            print(v, self.vertices[v].vecinos)
            
# Crear un grafo no dirigido
g = Grafo("0")

# Agregar vértices
g.agregarVertice("r")
g.agregarVertice("s")
g.agregarVertice("t")
g.agregarVertice("u")
g.agregarVertice("v")
g.agregarVertice("w")
g.agregarVertice("x")
g.agregarVertice("y")

# Agregar aristas
g.agregarArista("v", "r")
g.agregarArista("r", "s")
g.agregarArista("s", "w")
g.agregarArista("w", "t")
g.agregarArista("w", "x")
g.agregarArista("t", "x")
g.agregarArista("t", "u")
g.agregarArista("x", "u")
g.agregarArista("x", "y")
g.agregarArista("y", "u")

# Imprimir el grafo
g.imprimirGrafo()
```
