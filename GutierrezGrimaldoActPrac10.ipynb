```python
# Gutierrez Grimaldo Alejandro

# Estructura general que considera la lectura de un archivo de texto, el fallo de la apertura del archivo y el cierre del archivo
try:
    a=None
    a = open("Prueba1.txt",mode="r", encoding="utf-8")
except:
    print("El archivo no existe")
finally:
    if a is not None:
        a.close()

# Metodo read()
archivo = open("Prueba1.txt",mode="r", encoding="utf-8")
c1 = archivo.read(3)
c2 = archivo.read()
print(c1)
print(c2)
archivo.close()

# Metodo readline()
a3 = open("Prueba1.txt",mode="r", encoding="utf-8")
while True:
    linea = a3.readline()
    if not linea:
        break
    print(linea)
a3.close()

# Metodo readlines()
a4 = open("Prueba1.txt",mode="r", encoding="utf-8")
lista = a4.readlines()
numlin = 0
for linea in lista:
    numlin += 1
    print(f"{numlin} {linea}")
archivo.close()

# Sentencia with-as
with open("Prueba1.txt",mode="r", encoding="utf-8") as archivo:
    for linea in archivo:
        print(linea)
        
# Escritura de archivos
# Metodo write() y writelines()

cadena1 = 'Datos'
cadena2 = 'Secretos'
archivo2 = open("Datos2.txt",mode="w", encoding="utf-8")
print(cadena1 + '\n')
archivo2.write(cadena1 + '\n')
archivo2.writelines("hola")
archivo2.write(cadena2)
archivo2.close()

lista = ['lunes', 'martes', 'miercoles', 'jueves', 'viernes']
archivo2 = open("Datos2.txt",mode="w", encoding="utf-8")
archivo2.writelines(lista)
archivo2.close()

# Metodo seek()

archivo = open("Prueba1.txt",mode="r", encoding="utf-8")

archivo.seek(5)
cardenal = archivo.read(5)
print(cardenal)
print(archivo.tell())
archivo.close()

# Modulo Pickle y los metodos dump() y load()

import pickle
lista = ['algoritmos 1', 'algoritmos 2', 'estructuras']
archivo = open("materias.dat",mode="wb")
pickle.dump(lista,archivo)
archivo.close()
del lista
archivo = open("materias.dat",mode="rb")
lista = pickle.load(archivo)
print(lista)
archivo.close()

# Crear arbol de directorios o carpetas
import os
def crear_directorio(ruta):
    try:
        os.mkdir(ruta)
    except OSError:
        print("La creación del directorio %s falló" % ruta)
    else:
        print("Se ha creado el directorio: %s " % ruta)
    os.chdir(ruta)
crear_directorio('./')

# Recorrer un directorio
import os
rootDir = './'
for dirName, subdirList, fileList in os.walk(rootDir):
    print('Directorio encontrado: %s' % dirName)
    for fname in fileList:
        print('\t%s' % fname)
        
# Atributos de un archivo
archivo = open("Prueba1.txt",mode="r", encoding="utf-8")
contenido = archivo.read()
nombre = archivo.name
modo = archivo.mode
encoding = archivo.encoding
archivo.close()

if archivo.closed:
    print("El archivo se ha cerrado")
else:
    print("El archivo no se ha cerrado")
print(contenido)
print(nombre)
print(modo)
print(encoding)
```

```python
# Gutierrez Grimaldo Alejandro

import pathlib
import os

def menu():
    print('''Menu
          1) Agregar
          2) Mostrar
          3) Buscar
          4) Salir''')
    
def leerAgenda(agenda, nombreArchivo):
    if os.path.exists(nombreArchivo):
        archivo = open(nombreArchivo,mode="r", encoding="utf-8")
        for linea in archivo:
            contacto, telefono, email = linea.strip().split(";")
            agenda.setdefault(contacto, (telefono, email))
    else:
        with open(nombreArchivo, 'w') as archivo:
            pass
        
def agregar(agenda, nombreArchivo):
    print("Agregar contacto")
    nombre = input("Nombre: ")
    if agenda.get(nombre):
        print("El contacto ya existe")
    else:
        telefono = input("Telefono: ")
        email = input("Email: ")
        agenda.setdefault(nombre, (telefono, email))
        with open(nombreArchivo, 'a') as archivo:
            archivo.write(f"{nombre};{telefono};{email}\n")
        print("Contacto agregado con exito")
        
def mostraarContactos(agenda):
    print("Mis contactos")
    if len(agenda) > 0:
        for contacto, datos in agenda.items():
            print(f"Nombre: {contacto}")
            print(f"Telefono: {datos[0]}")
            print(f"Email: {datos[1]}")
            print()
    else:
        print("No hay contactos")
        
agenda = {}
nomArchivo = "agenda.txt"
leerAgenda(agenda, nomArchivo)

agregar(agenda, nomArchivo)
mostraarContactos(agenda)
```

```python
# Gutierrez Grimaldo Alejandro

import csv

with open('people.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    for row in reader:
        print(row)
        
csvFile.close()

# para no tener espacios en la salida despues del delimitador
csv.register_dialect('myDialect', delimiter = ',',  skipinitialspace=True)

with open('people.csv', 'r') as csvFile:
    reader = csv.reader(csvFile, dialect='myDialect')
    for row in reader:
        print(row)
        
csvFile.close()

# leer con otros delimitadores
csv.register_dialect('myDialect', delimiter = '|')

with open('myDialect.csv', 'r') as csvFile:
    reader = csv.reader(csvFile, dialect='myDialect')
    for row in reader:
        print(row)
        
# Lectura y almacenamiento en diccionario
with open('people.csv', 'r') as csvFile:
    reader = csv.DictReader(csvFile)
    for row in reader:
        print(dict(row))
        
csvFile.close()
```
