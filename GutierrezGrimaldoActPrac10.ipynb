```python
def encriptar_archivo(origen, destino):
    with open(origen, 'r') as file:
        texto = file.read()

    texto_encriptado = ""
    for caracter in texto:
        if 'A' <= caracter <= 'U' or 'a' <= caracter <= 'u':
            texto_encriptado += chr(ord(caracter) + 5)
        elif 'V' <= caracter <= 'Z' or 'v' <= caracter <= 'z':
            texto_encriptado += chr(ord(caracter) - 21)
        else:
            texto_encriptado += caracter

    with open(destino, 'wb') as file:  # Cambio a modo binario 'wb'
        file.write(texto_encriptado.encode('utf-8'))  # Codificar en utf-8 para escritura binaria

def descifrar_archivo(cifrado, destino):
    with open(cifrado, 'rb') as file:  # Modo de lectura binaria 'rb'
        texto_cifrado = file.read().decode('utf-8')  # Decodificar como utf-8

    texto_descifrado = ""
    for caracter in texto_cifrado:
        if 'A' <= caracter <= 'U' or 'a' <= caracter <= 'u':
            texto_descifrado += chr(ord(caracter) - 5)
        elif 'V' <= caracter <= 'Z' or 'v' <= caracter <= 'z':
            texto_descifrado += chr(ord(caracter) + 21)
        else:
            texto_descifrado += caracter

    with open(destino, 'w', encoding='utf-8') as file:  # Escribir como texto con codificación utf-8
        file.write(texto_descifrado)

# Usar la función para encriptar un archivo de texto
encriptar_archivo('archivo_original.txt', 'archivo_encriptado.txt')

# Usar la función para descifrar el archivo encriptado
descifrar_archivo('archivo_encriptado.txt', 'archivo_descifrado.txt')
```

# Actividad No.6

```python
import pickle

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1

class AVLTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        self.root = self._insert(self.root, key)

    def _insert(self, node, key):
        if not node:
            return Node(key)
        elif key < node.key:
            node.left = self._insert(node.left, key)
        else:
            node.right = self._insert(node.right, key)

        node.height = 1 + max(self._get_height(node.left), self._get_height(node.right))

        balance = self._get_balance(node)

        if balance > 1:
            if key < node.left.key:
                return self._rotate_right(node)
            else:
                node.left = self._rotate_left(node.left)
                return self._rotate_right(node)

        if balance < -1:
            if key > node.right.key:
                return self._rotate_left(node)
            else:
                node.right = self._rotate_right(node.right)
                return self._rotate_left(node)

        return node

    def _rotate_left(self, z):
        y = z.right
        T2 = y.left

        y.left = z
        z.right = T2

        z.height = 1 + max(self._get_height(z.left), self._get_height(z.right))
        y.height = 1 + max(self._get_height(y.left), self._get_height(y.right))

        return y

    def _rotate_right(self, z):
        y = z.left
        T3 = y.right

        y.right = z
        z.left = T3

        z.height = 1 + max(self._get_height(z.left), self._get_height(z.right))
        y.height = 1 + max(self._get_height(y.left), self._get_height(y.right))

        return y

    # Resto de las funciones de balanceo y otras necesarias

    def _get_height(self, node):
        if not node:
            return 0
        return node.height

    def _get_balance(self, node):
        if not node:
            return 0
        return self._get_height(node.left) - self._get_height(node.right)

    def guardar_arbol(self, nombre_archivo):
        with open(nombre_archivo, 'wb') as file:
            pickle.dump(self.root, file)

    def recuperar_arbol(self, nombre_archivo):
        with open(nombre_archivo, 'rb') as file:
            self.root = pickle.load(file)

    def imprimir_arbol(self, node, nivel=0):
        if node is not None:
            self.imprimir_arbol(node.right, nivel + 1)
            print("  " * nivel + str(node.key))
            self.imprimir_arbol(node.left, nivel + 1)

# Ejemplo de uso
tree = AVLTree()

# Agregar elementos al árbol
elementos = [10, 20, 30, 40, 50, 25]
for elemento in elementos:
    tree.insert(elemento)

# Guardar el árbol en un archivo binario
tree.guardar_arbol('arbol.bin')

# Recuperar el árbol desde el archivo binario
tree.recuperar_arbol('arbol.bin')

# Imprimir el árbol recuperado
print("Árbol AVL recuperado desde el archivo:")
tree.imprimir_arbol(tree.root)
```

```python
import matplotlib.pyplot as plt

# Lee el archivo y almacena los puntos en las listas x e y
x = []
y = []

with open('pares_puntos.txt', 'r') as file:
    for line in file:
        values = line.split()
        for i in range(0, len(values), 2):
            x.append(float(values[i]))
            y.append(float(values[i + 1]))

# Muestra los puntos en pantalla
print("Puntos:")
for i in range(len(x)):
    print(f"({x[i]}, {y[i]})")

# Grafica los puntos
plt.scatter(x, y)
plt.title('Gráfico de Puntos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```
